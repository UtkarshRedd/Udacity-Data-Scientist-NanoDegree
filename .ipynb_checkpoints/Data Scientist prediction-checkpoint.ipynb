```python
import pandas as pd
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r"C:\Users\utkar\Downloads\Udacity project Dataset\Stack Overflow\survey_results_public.csv")
data.info()
data.head()
```

```python
df1 = data[['DeveloperType', 'JobSatisfaction',
          'CareerSatisfaction', 'EducationTypes',
          'HaveWorkedFramework', 'HaveWorkedLanguage','HaveWorkedDatabase', 
            'HaveWorkedPlatform','FormalEducation', 'MajorUndergrad','Salary']]
df1 = df1.dropna(subset = ['DeveloperType']) #dropping null values in DeveloperType because we want to find values related to DataScientist. So if that value is null then we don't have any use for that row, and we can't replace it with another value too
df1 = df1.reset_index()
df1.head()
```

```python
sns.heatmap(df1.isnull())
```

Imputing missing values in the numerical columns by the mean. (Salary column will be dropped later)

```python
num_vars = df1.select_dtypes(include = ['float', 'int']).columns
#num_vars = num_vars.drop(['Salary'])
for col in num_vars:
        df1[col].fillna((df1[col].mean()), inplace=True)
```

```python
df1[['JobSatisfaction','CareerSatisfaction']].isnull().sum() #Checking to see if any null values
```

```python
df1['Y'] = 0 #Adding a column Y which will be 1 for Data Scientists and 0 for others. This will be the target variable
df1[['Y', 'DeveloperType']] 
```

```python
print(df1.columns.get_loc('Y')) #Finding the column index to substitute in df1.iat[] below
```

```python
#Creating a column Y, with value=1, for Data Scientists and 0 for others
flag = 0
for i in range(df1.shape[0]):
    sentence = df1['DeveloperType'][i].split(';')
    for word in sentence:
        word = word.strip()
        if word == "Data scientist":
            #print("hello ", i)
            df1.iat[i, 12] = 1
```

From here on, the basic idea is to create a table with columns of each categorical entity like 'Self-taught', 'Online course'.. 'Java', 'AngularJS' etc. Each of these will have a column with 0/1 values. From this table we can relate what all <b>labels</b> are related to a particular DeveloperType. We can understand what all a Data Scientist works with (in HaveWorkedFramework, HaveWorkedLanguage etc) or what is the education type of a DeveloperType.

```python
#Creating a new set of columns for every categorical entity with default value as 0
for cols in df1.columns:
    if cols not in ['DeveloperType','JobSatisfaction','CareerSatisfaction', 'Salary', 'index']:
        Blacklist = set()
        for sentences in df1[cols]:
            sentence = str(sentences).split(';')
            for word in sentence:
                if word != 'nan':
                    word = word.strip()
                    #print(word)
                    Blacklist.add(word)
        #print(Blacklist)
        for col in Blacklist:
            df1.insert(df1.shape[1], column = col, value = 0, allow_duplicates = False)
```

```python
df1 = df1.drop(['index','0','1'],axis = 1)
df1
```

```python
for cols in df1.columns: 
    if cols not in ['DeveloperType','JobSatisfaction','CareerSatisfaction', 'Salary', 'index']:
        for i in range(df1.shape[0]):
            sentence = str(df1[cols][i]).split(';')
            for word in sentence:
                word = word.strip()
                df1.at[i,word] = 1 #Putting values 1 for respective columns in the new table, where a particular DeveloperType has filled a value matching with that column name
```

```python
df1
```

```python
final_df = df1.drop(['nan', '0', '1'], axis = 1)
```

```python
final_df # This is the dataframe with all the categorical values represented as 0 or 1
```

```python
X = final_df.drop(['DeveloperType', 'JobSatisfaction', 'CareerSatisfaction', 'EducationTypes', 'HaveWorkedFramework',
                   'HaveWorkedLanguage', 'HaveWorkedDatabase', 'HaveWorkedPlatform', 'FormalEducation', 'MajorUndergrad','Salary'], 
                  axis = 1)
```

```python
X
```

```python
#Dividing into X1 and y. y is the target variable
y = X['Y']
X1 = X.drop(['Y'],axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size = .3, random_state = 42)
#Dividing into test and train data
```

```python
log_reg = LogisticRegression(max_iter = 10000)
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
print(accuracy_score(y_test, y_pred))
#Using logistic Regression because, it is based on the Sigmoid funtion, which performs well with Binary data like 0,1
#As you can see, the accuracy of prediction is 91%. This means that the model can predict whether a person filling the form is 
#a Data Scientist or not, based on the things he has filled in the columns.
```

<strong>Note:</strong> If the person is a Data Scientist according to the model, doesn't mean that he/she isn't a Web Developer or Mobile Developer or any other DeveloperType too. One person can be many things, according to the data given under DeveloperType.
